## Exemple d'import de données depuis un fichier CSV

```python
url="https://gist.githubusercontent.com/netj/8836201/raw/6f9306ad21398ea43cba4f7d537619d0e07d5ae3/iris.csv"
iris=pd.read_csv(url)
iris
```

## Exemple de manipulation de donnée

### Projection (sous-ensemble de colonnes, SELECT)

```python
sous_ensemble_iris = [['sepal.length','variety']]
sous_ensemble_iris
```

### Selection des données selon conditions (WHERE)

```python
# Filtrer les Setosa avec petal.length < 1.3
setosa_petite_taille = iris[(iris['variety'] == 'Setosa') & (iris['petal.length'] < 1.3)]
```

### Group by / faire des calculs pour des sous-ensemble de données

```python
iris.groupby('variety').size()
```

### Moyenne

```python
# Calculer la moyenne de la longueur des pétales par variété
moyennes = iris.groupby('variety')['petal length (cm)'].mean()
```
