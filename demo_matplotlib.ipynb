# Demo Matplotlib

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
import numpy as np

url="https://gist.githubusercontent.com/netj/8836201/raw/6f9306ad21398ea43cba4f7d537619d0e07d5ae3/iris.csv"
iris=pd.read_csv(url)
iris
```

```python
plt.plot(iris['sepal.length'], iris['sepal.width'],linestyle=' ',marker='o',color='blue')
plt.xlabel('sepal.length')
plt.ylabel('sepal.width')
```

```python
iris_setosa = iris[iris['variety'] == "Setosa"]
iris_virginica = iris[iris['variety'] == "Virginica"]
iris_versi = iris[iris['variety'] == "Versicolor"]

plt.plot(iris_setosa['sepal.length'], iris_setosa['sepal.width'],
         marker='o', color='b', linestyle='', label="Setosa")

plt.plot(iris_virginica['sepal.length'], iris_virginica['sepal.width'],
         marker='o', color='r', linestyle='', label="Virginica")

plt.plot(iris_versi['sepal.length'], iris_versi['sepal.width'],
         marker='o', color='g', linestyle='', label="Versicolor")

plt.legend()
```

```python
fig, axes - plt.subplots(2, 3, figsize=(10,10))

sub1 = axes[0,0]
sub1.plot(iris_setosa['sepal.length'], iris_setosa['sepal.width'], linestyle='', marker='o' )
sub1.set_ylabel('Sepal width')
sub1.set_title('Setosa')

sub2 = axes[0,1]
sub2.plot(iris_virginica['sepal.length'], iris_virginica['sepal.width'], linestyle='', marker='o')
sub2.set_title('Virginica')
sub2.set_xlabel('Sepal length')

sub3 = axes[0,2]
sub3.plot(iris_versi['sepal.length'], iris_versi['sepal.width' ], linestyle='', marker='o')
sub3.set_title('Versicolor')

sub4 = axes[1,0]
sub4.plot(iris_setosa['petal.length'], iris_setosa['petal.width'], linestyle='', marker='o')
sub4.set_ylabel('Petal width')

sub5 = axes[1,1]
sub5.plot(iris_virginica['petal.length'], iris_virginica['petal.width'], linestyle='', marker='o')
sub5.set_xlabel('Petal length')

sub6 = axes[1,2]
sub6.plot(iris_versi['petal.length'], iris_versi['petal.width'], linestyle='', marker='o')
```
